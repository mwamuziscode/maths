```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the number of atoms in each direction
n = 5
spacing = 1.0  # distance between atoms

# Generate coordinates for atoms in a simple cubic lattice
x, y, z = np.meshgrid(range(n), range(n), range(n))
x = x.flatten() * spacing
y = y.flatten() * spacing
z = z.flatten() * spacing

# Plot the lattice
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='cyan', s=100, edgecolor='k')

ax.set_title("Simple Cubic Crystal Lattice")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def bcc_lattice(n, spacing):
    atoms = []
    
    # Corner atoms
    for i in range(n):
        for j in range(n):
            for k in range(n):
                atoms.append((i * spacing, j * spacing, k * spacing))
                # Body-centered atom
                if i < n - 1 and j < n - 1 and k < n - 1:
                    center = ((i + 0.5) * spacing, (j + 0.5) * spacing, (k + 0.5) * spacing)
                    atoms.append(center)
    return np.array(atoms)

atoms = bcc_lattice(n=4, spacing=1.0)

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(atoms[:, 0], atoms[:, 1], atoms[:, 2], c='orange', s=100, edgecolor='k')
ax.set_title("Body-Centered Cubic (BCC) Lattice")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
def fcc_lattice(n, spacing):
    atoms = []
    
    # Corner and face atoms
    for i in range(n):
        for j in range(n):
            for k in range(n):
                base = np.array([i, j, k]) * spacing
                atoms.append(tuple(base))  # corner atom

                # Face-centered atoms
                if i < n - 1 and j < n - 1:
                    atoms.append(tuple(base + [0.5 * spacing, 0.5 * spacing, 0]))
                if i < n - 1 and k < n - 1:
                    atoms.append(tuple(base + [0.5 * spacing, 0, 0.5 * spacing]))
                if j < n - 1 and k < n - 1:
                    atoms.append(tuple(base + [0, 0.5 * spacing, 0.5 * spacing]))
    return np.array(atoms)

atoms = fcc_lattice(n=3, spacing=1.0)

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(atoms[:, 0], atoms[:, 1], atoms[:, 2], c='lime', s=100, edgecolor='k')
ax.set_title("Face-Centered Cubic (FCC) Lattice")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
